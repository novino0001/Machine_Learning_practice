```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=10, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)

```

```python
reg = linear_model.LinearRegression()
reg.fit(X,y)
print("coef is : ",reg.coef_)
print("intercept : ",reg.intercept_)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="green")
```

```python
print("m is :",reg.coef_)
print("intercept : ",reg.intercept_)
```

```python
learning_rate = 0.01
m = 48.53
b = 0
```

```python
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = learning_rate*loss_slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((m * X) + b).reshape(10)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="red",label = 'final')
plt.plot(X,y_pred1,color="green",label = 'steps')
```
